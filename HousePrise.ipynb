```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import norm
from sklearn.preprocessing import StandardScaler #Library umum untuk machine learning
from scipy import stats
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
df_train = pd.read_csv('HousePrise.csv')
```

```python
print("Skewness: %f" % df_train['last_sold_price'].skew())
print("Kurtosis: %f" % df_train['last_sold_price'].kurt())
```

```python
var = 'house_sqft'
data = pd.concat([df_train['last_sold_price'], df_train[var]], axis=1)
data.plot.scatter(x=var, y='last_sold_price', ylim=(0,800000));
```

```python
var = 'year_built'
data = pd.concat([df_train['last_sold_price'], df_train[var]], axis=1)
data.plot.scatter(x=var, y='last_sold_price', ylim=(0,800000));
```

```python
var = 'house_sqft'
data = pd.concat([df_train['last_sold_price'], df_train[var]], axis=1)
f, ax = plt.subplots(figsize=(8, 6))
fig = sns.boxplot(x=var, y="last_sold_price", data=data)
fig.axis(ymin=0, ymax=800000);
```

```python
var = 'year_built'
data = pd.concat([df_train['last_sold_price'], df_train[var]], axis=1)
f, ax = plt.subplots(figsize=(16, 8))
fig = sns.boxplot(x=var, y="last_sold_price", data=data)
fig.axis(ymin=0, ymax=800000);
plt.xticks(rotation=90);
```

```python
sns.set()
cols = ['house_sqft', 'num_of_bedrooms', 'num_of_bathrooms', 'year_built', 'tax_assessed_value', 'last_sold_price', 'rate_per_sqfoot', 'city', 'home_type', 'school_rating_1to10']
sns.pairplot(df_train[cols], size = 2.5)
plt.show();
```
